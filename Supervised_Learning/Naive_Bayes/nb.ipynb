```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

!curl "http://www.codeheroku.com/static/workshop/datasets/titanic.csv" -o "titanic.csv"
```

```python
df = pd.read_csv("titanic.csv").drop(["Name"],axis=1)
df.head()
df.describe()
```

```python
ax2 = df.plot.scatter(x='Age',y='Fare',c='Survived',colormap='bwr')
```

```python
x = df.drop(['Survived'],axis=1)
y = df['Survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
x_train.shape
```

```python
model = GaussianNB().fit(x_train,y_train)
```

```python
predicted_values = model.predict(x_test)
```

```python
for actual,predicted in zip(y_test.values,predicted_values):
    print(actual,predicted)
```

```python
model.score(x_test,y_test)
```
